```python
comment = "ไฟถนนดับหมดทั้งซอย มืดและอันตรายมาก"

prompt = f"""
You are an urgency-classification assistant.
Analyze the Thai comment and return only JSON with:
- urgency_score (0–1)
- urgency_level (low/medium/high)
- reasons

Comment:
{comment}
"""

response = client.models.generate_content(
    model="gemini-2.0-flash",
    contents=prompt,
    config=types.GenerateContentConfig(
        response_mime_type="application/json"
    )
)

print(response.text)
```
